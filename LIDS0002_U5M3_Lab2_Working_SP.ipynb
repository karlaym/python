### Problema 7

##### ¿Qué escuelas tienen una Asistencia promedio de estudiantes menor a 70 %?

```python
%sql select name_of_school as name, Average_Student_Attendance as average from SCHOOLS where Average_Student_Attendance<70
```

Haga doble clic __aquí__ para obtener una pista. 

<!--
El tipo de datos de la columna "Average_Student_Attendance" es varchar.
Por lo tanto, no puede usarlo tal como está en la cláusula where para una comparación numérica.
Primero use la función CAST () para convertirla (cast) a DECIMAL o DOBLE
p.ej. CAST ("Nombre_Columna" as DOUBLE)
o simplemente: DECIMAL ("Column_Name")
-->

Haga doble clic aquí para obtener otra pista. 

<!--
No olvides que el signo de edad '%' debe eliminarse antes de convertirse (casting)
-->

Haga doble clic __aquí__ para ver la solución. 

<!-- Solución:

%sql SELECT "Name_of_School", "Average_Student_Attendance"  \
     from SCHOOLS \
     where CAST ( REPLACE("Average_Student_Attendance", '%', '') AS DOUBLE ) < 70 \
     order by "Average_Student_Attendance"
     
o

%sql SELECT "Name_of_School", "Average_Student_Attendance"  \
     from SCHOOLS \
     where DECIMAL ( REPLACE("Average_Student_Attendance", '%', '') ) < 70 \
     order by "Average_Student_Attendance"

-->
